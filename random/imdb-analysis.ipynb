# Analysis of my own IMDb data

#### Various interesting graphs have been made from IMDb data (based on 400+ user ratings)

```python
import pandas as pd
from matplotlib import pyplot as plt
```

#### Data can be downloaded in csv format from IMDb

```python
# Requires different encoding ('UTF-8' is default)

ratings = pd.read_csv("ratings.csv", encoding = "ISO-8859-1")
```

```python
ratings.head()
```

```python
# Remove useless columns

del ratings['URL']
del ratings['Const']
```

```python
ratings
```

```python
# Convert columns to the correct type
to_change = ['Your Rating', 'IMDb Rating', 'Runtime (mins)', 'Year', 'Num Votes']

for i in to_change:
    ratings[i] = pd.to_numeric(ratings[i])
```

```python
ratings
```

```python
ratings.sort_values(by=['Year'])
```

## Graphs

```python
years = ratings['Year']
years.shape
```

```python
dic_years = {}

for i in years:
    if i not in dic_years:
        dic_years[i] = 1
    else:
        dic_years[i] += 1
        
```

```python
X = []
Y = []

for i in dic_years:
    X.append(i)
    Y.append(dic_years[i])
```

```python
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.bar(X, Y)
plt.show()
```

```python
rating = ratings['IMDb Rating']
dic_rating = {}

for i in rating:
    if i not in dic_rating:
        dic_rating[i] = 1
    else:
        dic_rating[i] += 1
```

```python
X_rating, Y_rating = [], []

for i in dic_rating:
    X_rating.append(i)
    Y_rating.append(dic_rating[i])
```

```python
plt.scatter(X_rating, Y_rating)
plt.xlabel('IMDb Rating')
plt.ylabel('Number of Movies')
plt.show()
```

```python
runtime = ratings['Runtime (mins)']
runtime
```

```python
dic_runtime = {}

for i in runtime:
    if i not in dic_runtime:
        dic_runtime[i] = 1
    else:
        dic_runtime[i] += 1

X_runtime, Y_runtime = [], []

for i in dic_runtime:
    X_runtime.append(i)
    Y_runtime.append(dic_runtime[i])
```

```python
plt.bar(X_runtime, Y_runtime)
plt.xlabel('Runtime in Minutes')
plt.ylabel('Number of Movies')
plt.show()
```

```python
myrating = ratings['Your Rating']
```

```python
dic_myrating = {}

for i in myrating:
    if i not in dic_myrating:
        dic_myrating[i] = 1
    else:
        dic_myrating[i] += 1

X_myrating, Y_myrating = [], []

for i in dic_myrating:
    X_myrating.append(i)
    Y_myrating.append(dic_myrating[i])
```

```python
plt.bar(X_myrating, Y_myrating)
plt.xlabel('My Rating')
plt.ylabel('Number of Movies')
plt.show()
```
